# Punto Matplotlib

Construir una cuadrícula de 9 celdas en donde grafique a la función 

$$ f(x) = \frac{\sin nx}{x} $$

en el dominio de $[-2\pi, 2\pi]$ y para $n \in \{ 1,2,3,4,5,,6,7,8,9 \}$

1. A cada celda colocarle por título $$ f(x) = \frac{\sin nx}{x} $$ para el $n$ correspondiente a la celda.
2. Denotar los ejes con $x$ para el eje horizontal y $f(x)$ para el eje verical. 
3. Colocarle cuadrícula solo a la última celda. 


```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
x = np.arange(-2*np.pi, 2*np.pi,0.1)
```

```python
plt.figure(figsize = (17,17))

for n in range(1,10):
    plt.subplot(3,3,n)
    plt.plot(x,np.sin(n*x)/x, label = 'n = '+str(n))
    plt.title(r'$f(x) = \frac{\sin(nx)}{x}$')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend() 
    if n == 9:
        plt.grid(True)


plt.savefig('comportamiento.jpg')
plt.show() 
```
