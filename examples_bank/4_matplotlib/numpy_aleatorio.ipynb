# Dos generadores aleatorios de numpy

```python
import numpy as np 
import matplotlib.pyplot as plt 
n = 10000000                     # NÃºmero de muestras 
x = np.random.rand(n)
plt.hist(x, bins = 50 )
plt.show()
```

```python
import numpy as np 
import matplotlib.pyplot as plt 
x = np.random.randn(10000000)
plt.hist(x, bins = 500 )
plt.show()
```
